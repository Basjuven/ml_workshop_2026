## 1. What is Machine Learning?

In traditional programming, we write **explicit rules** so that the computer can solve a problem.  
In *Machine Learning* (ML), instead of writing all the rules, we give the computer **data** and **examples**, and the system **learns patterns** by itself.

### Simple definition

> **Machine Learning** is a set of methods that allow a computer to *learn* from data in order to make predictions or decisions, without being explicitly programmed for every case.

### Everyday examples of ML

- YouTube/Netflix recommendations  
- Spam filters in email  
- Weather prediction  
- Navigation apps (route selection)  
- Facial recognition in photos  
- Bank fraud detection

## 2. Traditional programming vs. Machine Learning

### Traditional programming (rule-based)

In classical programming:

- We define **rules** using `if`, `else`, `for`, etc.
- The computer follows those rules exactly.
- If the rules do not consider a case, the program fails.

Example: classifying whether a number is even or odd.


```python
# Example (explicit rules): even or odd

number = int(input("Enter a number from 1 to 100:" ))

if number % 2 == 0:
    print("It is even")
else:
    print("It is odd")
```

### Machine Learning (data-based)

In ML:

- We do not write all the rules manually.
- We provide **labeled data** (examples with correct answers).
- The algorithm learns patterns.
- The trained model can then make predictions on new data.

Example: predicting whether a student will pass or not, using attendance, study hours, assignments, etc.  
It is difficult to write exact rules by hand for every case, but a model can learn from many examples.

## 4. Main types of Machine Learning

Although there are many algorithms, at an introductory level we will focus on two groups:

### 4.1 Supervised learning
- The data come with **labels** (correct answers).
- The model learns to predict the label.

Examples:
- Predicting **daily sales** for a store (regression).
- Classifying emails as **spam/not spam** (classification).
- Determining whether an image contains a **cat or a dog**.

### 4.2 Unsupervised learning
- The data **do not have labels**.
- The model looks for **hidden patterns** or **groups** (clusters).

Examples:
- Grouping customers by shopping behavior.
- Grouping songs by similarity.
- Finding groups of students with similar habits (without using grades).

> In this course, we will work mainly with **supervised learning**, because it is more intuitive for beginners and easier to bring into the classroom.

```python
# Activity 2 (Interactive Quiz)

examples = {
    1: "A system that predicts tomorrow's electricity consumption using historical data.",
    2: "A system that groups store customers according to their behavior, without previous labels.",
    3: "A filter that decides whether a message is spam or not spam.",
    4: "An algorithm that groups songs by similarity without being told the genres."
}

print("Classify each example as:")
print("  S = Supervised")
print("  U = Unsupervised\n")

user_answers = {}
for i, text in examples.items():
    print(f"Example {i}: {text}")
    answer = input("Is it S or U? ").strip().upper()
    user_answers[i] = answer
    print()

# Correct answers
correct_answers = {1: "S", 2: "U", 3: "S", 4: "U"}

print("Results:")
correct = 0
for i in examples.keys():
    is_correct = user_answers[i] == correct_answers[i]
    status = "Correct" if is_correct else f"Incorrect (correct answer: {correct_answers[i]})"
    print(f"Example {i}: {status}")
    if is_correct:
        correct += 1

print(f"\nFinal score: {correct} out of {len(examples)}")
```
